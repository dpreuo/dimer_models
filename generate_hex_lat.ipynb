```python
n = 70

lattice0 = generate_lattice(uniform(n))
dim1 = gu.dimerise(lattice0)
lattice1 = gu.dimer_collapse(lattice0, dim1)
dim2 = gu.dimerise(lattice1)

out = gu.vertices_to_polygon(lattice1)
```

```python
dim3 = list(dim2) + [0] * (out.n_edges - lattice1.n_edges)
dim3 = np.array(dim3)
dim3.shape
out.n_edges
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10, 3), dpi=300)

ax = ax.flatten()
for a in ax:
    a.set_xticks([])
    a.set_yticks([])

pl.plot_edges(lattice0, labels=dim1, ax=ax[0])
pl.plot_edges(lattice1, labels=dim2, ax=ax[1])
pl.plot_edges(out, labels=dim3, ax=ax[2])
```
